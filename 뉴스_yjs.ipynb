```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib.font_manager as fm
from konlpy.tag import Komoran # 명사만 분리하기

# 한글 폰트 설정
font_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"

# 샘플 텍스트
text = """
지난해 말 이스라엘이 헤즈볼라와 관련된 한 금융 그룹을 공습하면서 헤즈볼라의 자금 출처를 둘러싼 의혹이 새롭게 커지고 있다.

헤즈볼라는 레바논에서 가장 강력한 무장 세력이자, 영향력 있는 시아파 이슬람교 정당으로, 레바논 의회 및 정부에서의 존재감도 상당하다. 이스라엘과 많은 서방 국가에서는 테러 단체로 지정되어 있다.

헤즈볼라는 1975~1990년 레바논 내전 당시 레바논 남부를 점령한 이스라엘에 반대하던 1980년대에 두각을 나타내기 시작했다.

건물에서 연기와 화염이 치솟는 모습사진 출처,EPA
사진 설명,2024년 11월 26일 이스라엘은 AQAH 금융 그룹의 레바논 남부 도시 시돈 소재 지부를 공습했다
지난해 10월, 이스라엘 방위군(IDF)은 이란의 지원을 받는 헤즈볼라의 자금 조달용 위장 조직이라면서 금융 기관 '알-카드 알-하산(AQAH)'의 여러 지부를 공습했다.

그러나 AQAH 측은 이러한 의혹을 부인하며 자신들은 민간 소액 대출 업체라고 주장했다.

‘알-카드 알-하산’이라고 적힌 문서가 바닥에 널브러진 모습사진 출처,Reuters
사진 설명,이스라엘의 공습으로 여러 AQAH 지부가 타격을 입은 다음 날인 2024년 10월 21일, 레바논 아크비예 소재 AQAH 지부 건물 밖에 문서가 나뒹굴고 있다
지난해 레바논 내에서 헤즈볼라와 이스라엘이 약 2개월간 대립하는 과정에서 헤즈볼라의 영향력이 큰 지역은 광범위한 피해를 보았다.

이 갈등은 그해 11월 27일에 발효된 휴전 협정으로 마무리되었으며, 이후 이웃국 시리아의 알-아사드 정권이 붕괴했다.

이 같은 상황은 헤즈볼라의 자금 조달 능력과 군사 장비 확보 능력에 큰 영향을 미칠 수 있다.

BBC 아랍어 서비스는 헤즈볼라의 핵심 수입원 및 레바논과 전 세계에 뻗어 있는 헤즈볼라의 금융 네트워크에 대해 살펴봤다.

이란과의 연관성
2024년 9월 21일 테헤란에서 열린 1980년 이라크 침공 기념 연례 군사퍼레이드에서 포착된 이란 IRGC 병사들사진 출처,EPA
사진 설명,미국은 이란 이슬람혁명수비대(IRGC)가 헤즈볼라에 매년 수억달러를 지원한다고 추정한다
2022년 미국 국무부는 이란이 헤즈볼라에 매년 자금 7억달러(약 1조원)를 제공했다고 추정한 바 있다.

2024년 9월 이스라엘의 공습으로 사망한 헤즈볼라의 지도자 하산 나스랄라는 2016년 연설에서 이란이 핵심 자금 원천이라고 주장하긴 했으나, 구체적인 수치를 제시하지는 않았다.

"우리의 예산, 급여, 지출, 식량, 물, 무기, 미사일은 모두 이란에서 나온다"는 설명이다.

이란 당국은 1980년대 레바논 그룹 창설을 도운 '이슬람혁명수비대(IRGC)'를 통해 헤즈볼라에 자금을 지원한다. 아울러 IRGC는 첨단 미사일, 드론 등 헤즈볼라의 주요 무기 공급원이기도 하다.

맑고 푸른 하늘을 가르는 하얀 연기사진 출처,Reuters
사진 설명,헤즈볼라와 이스라엘이 국경 간 적대 행위를 이어가던 2024년 10월 20일, 이스라엘의 '아이언 돔' 시스템이 레바논이 이스라엘을 향해 발사한 로켓포를 요격하는 모습
이렇듯 과거에는 이란의 지원에 대부분 의존했을 수도 있으나, 미국의 싱크탱크 '워싱턴 근동 정책 연구소'의 하닌 가다르 선임연구원은 현재는 헤즈볼라가 자금 출처를 다양화했다고 설명했다.

"대이란 제재로 인해 이란이 헤즈볼라에 예전처럼 많은 자금을 지원할 수 없게 되었기 때문"이다.

책 '헤즈볼라: 레바논 내 신의 당의 세계적 발자취'의 저자인 매튜 레빗에 따르면 헤즈볼라는 지금 세탁을 포함한 여러 불법적인 금융 활동을 통해 수입을 보충하고 있다.
"""

# data = [t for t in text if bool(t.strip())]
# print(data)
komoran = Komoran()
nouns = komoran.nouns(text)
nouns_text = " ".join(nouns)
print(nouns_text)


# 워드클라우드
wordcloud = WordCloud(font_path = font_path, width=800, height=400, background_color='white').generate(nouns_text)

# 시각화
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off') # 축 끄기
plt.savefig("word2.png") # 이미지로 저장
plt.show() # 제일 많이 사용하는 단어가 제일 크게 나옴
```
